```python
def sampl_frm_lat(mean,sigma,latent_dim):
    
    z_mean,z_sigma=mean,sigma
    
    epsilon=K.random_normal(shape=(K.shape(z_mean)[0],latent_dim),mean=0.,stddev=1.)
    
    lat= z_mean+z_sigma*epsilon
    
    return lat
    
```
